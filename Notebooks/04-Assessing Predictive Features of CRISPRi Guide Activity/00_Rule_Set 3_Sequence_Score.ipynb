```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from rs3.seq import predict_seq
import matplotlib
import gpplot as gpp
gpp.set_aesthetics(context = 'paper')
```

```python
figure3_outpath = '../../Figures/Figure 3'
```

```python
file_path = '../../Data/cleanedData/'
NewDatasetCombine = pd.read_csv(file_path + 'CombinedCleanedDatasets.csv', low_memory=False)
NewDatasetCombine
```

```python
#check for any Us in context sequences
any('u' in seq for seq in NewDatasetCombine['sgRNA Context Sequence'])
```

```python
any('U' in seq for seq in NewDatasetCombine['sgRNA Context Sequence'])
```

```python
#convert entire sequence to uppercase
def to_upper_case(seq_list):
    return ''.join([seq.upper() for seq in seq_list])

#convert all context sequences to all uppercase letters
NewDatasetCombine['sgRNA Context Sequence'] = NewDatasetCombine['sgRNA Context Sequence'].apply(to_upper_case)
```

```python
NewDatasetCombine.DataSet.value_counts()
```

```python
#calculate Rule Set 3 Sequence Scores for every context sequence with Chen tracrRNA
NewDatasetCombine['rs3ChenSeqScore'] =  predict_seq(NewDatasetCombine['sgRNA Context Sequence'], sequence_tracr='Chen2013')
```

```python
#create column which uniquely identifies the 6 datasets
NewDatasetCombine['cellDomainDataSet'] = NewDatasetCombine[['condition', 'Domain', 'DataSet']].apply(
    lambda x: "_".join(x.astype(str)), axis=1)
```

```python
#create one column for z-score regardless of which dataset it comes from / how z-score was calcualted
NewDatasetCombine['z_score'] = NewDatasetCombine['z_scored_avg_lfc'].combine_first(NewDatasetCombine['zscoreRelativeNC_signed']).combine_first(NewDatasetCombine['z-score relative to NC'])
```

```python
#calculate within-gene spearman correlation between RS3 sequence score and z-score
withinGeneCorr = NewDatasetCombine.groupby(['cellDomainDataSet',
                                            'Target Gene Symbol'])[['z_score','rs3ChenSeqScore']].corr('spearman').unstack().iloc[:, 1].reset_index()
withinGeneCorr.head()
```

```python
withinGeneCorr.columns = ['cellDomainDataSet', 'Target Gene Symbol', 'spearman correlation']
```

```python
withinGeneCorr.cellDomainDataSet.value_counts()
```

```python
# identify conditions for which spearman correlation is negative (i.e. viability screens)
viability_screens = (withinGeneCorr.cellDomainDataSet == 'K562_Kox1_Nunez') | \
(withinGeneCorr.cellDomainDataSet == 'A549_Kox1_InHouse') | \
(withinGeneCorr.cellDomainDataSet == 'A549_Zim3_InHouse') | \
(withinGeneCorr.cellDomainDataSet == 'HCT116_Kox1_InHouse') | \
(withinGeneCorr.cellDomainDataSet == 'HCT116_Zim3_InHouse') 

# sign adjust the spearman correlation for these screens such that more positive indicates stronger agreement
withinGeneCorr.loc[viability_screens, 'spearman correlation'] *= -1
```

```python
#plot distribution of within-gene spearman correlations in each of the 6 datasets
plt.figure(figsize=(8, 6))
sns.boxplot(x='cellDomainDataSet', y='spearman correlation', 
            data=withinGeneCorr, boxprops=dict(facecolor='lightgrey', edgecolor='black'),
                 whiskerprops=dict(color='black'),
                 capprops=dict(color='black'),
                 medianprops=dict(color='black'), whis = [5,95])
plt.title('RS3 Sequence Performance',fontsize=18)
plt.xlabel('')
plt.ylabel('Within-gene Spearman Correlation',fontsize=16)
plt.tight_layout()

plt.xticks(fontsize=16, rotation=90)
plt.yticks(fontsize=16)
sns.despine()
```

```python
#average z-scores across in-house datasets since we previously saw high correlation between them
InhouseEss = NewDatasetCombine[NewDatasetCombine['DataSet'] == 'InHouse']
other = NewDatasetCombine[NewDatasetCombine['DataSet'] != 'InHouse']
InhouseEssAvg = InhouseEss.groupby(['sgRNA Context Sequence','rs3ChenSeqScore','Target Gene Symbol']).agg(z_score = ('z_score','mean'),
                                                                                                          num_datasets = ('cellDomainDataSet','nunique')).reset_index()
InhouseEssAvg = InhouseEssAvg[InhouseEssAvg['num_datasets'] == 4]
InhouseEssAvg['cellDomainDataSet'] = 'This Study'

#repeat within-gene spearman correlation with averaged in-house data
newData = pd.concat([InhouseEssAvg,other])
withinGeneCorr = newData.groupby(['cellDomainDataSet','Target Gene Symbol'])[['z_score', 'rs3ChenSeqScore']].corr('spearman').unstack().iloc[:, 1]
withinGeneCorr = withinGeneCorr.reset_index()
withinGeneCorr.columns = ['cellDomainDataSet', 'Target Gene Symbol', 'spearman correlation']

viability_screens = (withinGeneCorr.cellDomainDataSet == 'K562_Kox1_Nunez') | \
(withinGeneCorr.cellDomainDataSet == 'This Study')
withinGeneCorr.loc[viability_screens, 'spearman correlation'] *= -1

#recreate DataSet column to use for plotting
withinGeneCorr['DataSet'] = np.where(withinGeneCorr['cellDomainDataSet'] == 'This Study',
                                     'This Study', withinGeneCorr['cellDomainDataSet'].str.split('_',expand=True)[2])
withinGeneCorr
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='DataSet', y='spearman correlation', 
            data=withinGeneCorr, boxprops=dict(facecolor='lightgrey', edgecolor='black'),
                 whiskerprops=dict(color='black'),  # Whisker line color
                 capprops=dict(color='black'),  # Cap line color
                 medianprops=dict(color='black'), whis = [5,95], order = ['This Study', 'Nunez', 'Gilbert'])
plt.title('RS3 Sequence Performance',fontsize=18)
plt.xlabel('')
plt.ylabel('Within-gene Spearman Correlation',fontsize=16)
plt.tight_layout()

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
sns.despine()

gpp.savefig(figure3_outpath + "RS3seq_within_gene_spearmanr_boxplots_inhouse_averaged.pdf", dpi=300, bbox_inches='tight')
```
